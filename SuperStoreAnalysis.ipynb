# Import Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
data_file=pd.read_csv("Sample _Superstore.csv",encoding='latin-1')
```

```python
# Managing Date Formate
data_file['Order Date']=pd.to_datetime(data_file['Order Date'])
data_file['Ship Date']=pd.to_datetime(data_file['Ship Date'])
    
```

```python
# Add new columns for year,month and time_of_delivery
```

```python
data_file['Order Year']= data_file['Order Date'].dt.year
data_file['Order Month']= data_file['Order Date'].dt.month
data_file['Order Day Of Week']=data_file['Order Date'].dt.dayofweek
```

```python
data_file.head()
```

# Analysis and Visulization

## Monthly Sales of Store

```python
Monthly_sales= data_file.groupby('Order Month')['Sales'].sum()
sns.lineplot(Monthly_sales,marker='o')
plt.title("Monthly Sales Analysis")
plt.grid()
plt.xticks(range(1,13))
plt.show()
```

### Conclusion: November has the highest sales and February has the lowest sales.

## Sales Based On Product Category and Sub-Category

```python
# Q2. Analyze sales based on product categories and determine which category has the lowest sales and which category has the highest sales.
fig= sns.countplot(data_file,x='Category')
fig.bar_label(fig.containers[0])
plt.title('Main Product Categories')
plt.show()
```

```python
categories=data_file.groupby('Category')['Sales'].sum()
plt.pie(categories,labels=categories.index,autopct='%1.2f%%')
plt.title("Sales Based On Product")
plt.show()
```

### Conclusion: Category->Technology has the highest sales with 36.40% and Category-> Office supplies has the lowest sales with 31.30% just 1% less than Furniture Category.

```python
sns.countplot(data_file,x='Sub-Category')
plt.xticks(rotation=70)
plt.title("Sub_Categories Count")
plt.show()
```

```python
sub_cat_sales=data_file.groupby('Sub-Category')['Sales'].sum()
sns.lineplot(sub_cat_sales,marker='o')
plt.xticks(rotation=80)
plt.title('Total Sales Based On Sub Catagories')
plt.grid()
plt.show()
```

### conclusion: Chairs and Phones are the highest sale sub-category products.

## Monthly Profit Analysis

```python
monthly_profit=data_file.groupby('Order Month')['Profit'].sum()
sns.barplot(monthly_profit)
plt.title("Total Monthly Profit Chart")
plt.xlabel("Months")
plt.show()
```

### Conclusion: From the bar graph we conclude that month december has the highest profit gained and january has the lowest profit.

# Profit Analysis By Category and Sub-Category

```python
profit_by_category=data_file.groupby('Category')['Profit'].sum()
plt.pie(profit_by_category,labels=profit_by_category.index,autopct='%1.2f%%')
plt.title("Profit Analysis By Category")
plt.legend()
plt.show()
```

### Conclusion: Technology gained the highest profit with 50.79% and Furniture has the lowest contribution on profit with just 6.44%.

```python
profit_by_subcategory=data_file.groupby('Sub-Category')['Profit'].sum()
sns.lineplot(profit_by_subcategory,marker='o')
plt.title("Profit Analysis By Sub-Category")
plt.xticks(rotation=80)
plt.grid()
plt.show()
```

### Conclusion: Copiers has the highest profit and Tables is in loss.

# Analyze the sales and profit by customer segment.

```python
customers = data_file.groupby('Segment').agg(sales=('Sales', 'sum'), profit=('Profit', 'sum')).reset_index()
# Convert to long format using `melt()` for paired bars
customers_melted = customers.melt(id_vars='Segment', var_name='Metric', value_name='Amount')
sns.barplot(x='Segment', y='Amount', hue='Metric', data=customers_melted, palette=['blue', 'green'])
plt.xlabel("Customer Segment")
plt.ylabel("Total Amount")
plt.title("Sales & Profit by Customer Segment")
plt.legend(title="Metric")
plt.show()
```

### Conclusion: Consumer has the highest sales along with highest profit and Home Office has the lowwest sales with lowest profit.
